```python
import pandas as pd
import numpy as np
import yaml

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:,.2f}'.format)

with open('../../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

#Calling Dataset
file_name = "Adjusted.xlsx"
sheet_name = 'Dataset_for_Model'
file_path = f"{Setting['Output_Path_Ajusted']}/{file_name}"
Dataset = pd.read_excel(file_path,sheet_name = sheet_name)


#Finding the Share of Alpha and Beta in Coub-Dogllas production funcation
alpha_labour = Dataset["Wage"].sum() / Dataset["Output_Value"].sum()
beta_capital = Dataset["Capital"].sum() / Dataset["Output_Value"].sum()

#Finding the Productivity by year and industry
Dataset["Log_Productivity"] = (
    np.log(Dataset["Output_Value"])
    - beta_capital * np.log(Dataset["Capital"])
    - alpha_labour * np.log(Dataset["Wage"])
)


#Finding the Productivity by year
Productivity_Dataset = (
    Dataset.groupby(["Industry_Code"], as_index=False)
      .agg(
          Log_Industry_Effect=("Log_Productivity", "mean")
      )
)




output_file_name = 'Adjusted.xlsx'
sheet_name= 'Dataset_for_Model'
path = f"{Setting['Output_Path_Ajusted']}/{output_file_name}"
with pd.ExcelWriter(path, engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    Dataset.to_excel(writer,sheet_name=sheet_name ,index=False)
```
