```python
Dataset = pd.merge(Dataset, grouped_expanded, on='Industry_Code', how='left')
Dataset.drop(columns=['Industry_Name'], inplace=True)
Dataset = Dataset.rename(columns={'Industry_New_Name': 'Industry_Name'})
cols = Dataset.columns.tolist()
cols.insert(3, cols.pop(cols.index('Industry_Name')))
Dataset = Dataset[cols]
WorkShop_Codes = Dataset.iloc[:,1:4].drop_duplicates()
```
