```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1399 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T06 = pd.read_excel(file_path, sheet_name='T06', skiprows=1)
        T06 = T06.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوئیل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T06 = T06[T06['Year'] != 0]
        Data.append(T06)
    
    elif year == 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T06 = pd.read_excel(file_path, sheet_name='T06', skiprows=1)
        T06 = T06.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوییل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T06 = T06[T06['Year'] != 0] 
        Data.append(T06)    
    
    elif year == 1401 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T06 = pd.read_excel(file_path, sheet_name='T06', skiprows=2)
        T06 = T06.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوییل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T06 = T06[T06['Year'] != 0] 
        Data.append(T06)
        
Dataset = pd.concat(Data, ignore_index=True)
Dataset = Dataset.dropna()
```

```python
output_file_name = '9_WorkShop_Input_Energy_By_Activity.xlsx'
Dataset.to_excel(f"{Setting['Output_Path']}/{output_file_name}", index=False)
```
