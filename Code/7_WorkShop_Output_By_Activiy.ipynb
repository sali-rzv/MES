```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1399 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T05 = pd.read_excel(file_path, sheet_name='T05', skiprows=1)
        T05 = T05.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Output_Value',
            'محصولات تولید شده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'ارزش غذای طبخ شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته شده' : 'Output_Value_Electricity',
            'آب فروخته شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
            'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
            'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به فروش رسیده اند' : 'Output_Value_peculate'
        })

        T05 = T05[T05['Year'] != 0] 
        Data.append(T05)
    
    elif year == 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T05 = pd.read_excel(file_path, sheet_name='T05', skiprows=1)
        T05 = T05.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Output_Value',
            'محصولات تولیدشده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'غذای طبخ‌شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته‌شده' : 'Output_Value_Electricity',
            'آب فروخته‌شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
            'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه‌ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
             'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به‌فروش‌رسیده‌اند' : 'Output_Value_peculate'
        })

        T05 = T05[T05['Year'] != 0] 
        Data.append(T05)    

    elif year == 1401 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T05 = pd.read_excel(file_path, sheet_name='T05', skiprows=2)
        T05 = T05.rename(columns={
            'سال':'Year',
            'سطح کد فعالیت':'Industry_Category_Code',
            'کد فعالیت':'Industry_Code',
            ' فعالیت':'Industry_Name',
            'جمع' : 'Output_Value',
            'محصولات تولیدشده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'غذای طبخ‌شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته‌شده' : 'Output_Value_Electricity',
            'آب فروخته‌شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
           'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه‌ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
            'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به‌فروش‌رسیده‌اند' : 'Output_Value_peculate'
        })

        T05 = T05[T05['Year'] != 0] 
        Data.append(T05)
    
Dataset = pd.concat(Data, ignore_index=True)
```

```python
output_file_name = '7_WorkShop_Output_By_Activity.xlsx'
Dataset.to_excel(f"{Setting['Output_Path']}/{output_file_name}", index=False)
```
