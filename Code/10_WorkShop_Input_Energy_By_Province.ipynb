```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1399 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T16 = pd.read_excel(file_path, sheet_name='T16', skiprows=1)
        T16 = T16.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوئیل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T16 = T16[T16['Year'] != 0] 
        Data.append(T16)
    
    elif year == 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T16 = pd.read_excel(file_path, sheet_name='T16', skiprows=1)
        T16 = T16.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوییل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T16 = T16[T16['Year'] != 0]
        Data.append(T16)    
    
    elif year == 1401 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T16 = pd.read_excel(file_path, sheet_name='T16', skiprows=2)
        T16 = T16.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            'جمع' : 'Input_Energy_Value',
            'نفت سفید' : 'Kerosene',
            'نفت سیاه و نفت کوره' : 'Fueloil',
            'بنزین' : 'Gasoline' ,
            'گاز طبیعی' : 'Natural_Gas' , 
            'گاز مایع ' : 'LNG' ,
            'گازوییل' : 'Gasoil' ,
            'آب' : 'Water' ,
            'برق' : 'Electricity' ,
            'سایر مواد سوختی' : 'Other_Fuels',
            'اتان' : 'Ethan',
            'زغال چوب' : 'Charcoal',
            'زغال سنگ' :'Coal'
        })

        T16 = T16[T16['Year'] != 0] 
        Data.append(T16)
    
    
Dataset = pd.concat(Data, ignore_index=True)
```

```python
Dataset["Province"] = Dataset["Province"].replace({
"چهارمحال وبختيارئ" : "چهارمحال و بختیاری",
"اردبيل" : "اردبیل",
"ايلام" : "ایلام",
"آذربايجان شرقي" : "آذربایجان شرقی",
"آذربايجان غربي" : "آذربایجان غربی",
"خراسان جنوبي" : "خراسان جنوبی",
"خراسان رضوئ" : "خراسان رضوی",
"خراسان شمالي" : "خراسان شمالی" ,
"سيستان وبلوچستان" : "سیستان و بلوچستان" , 
"قزوين" : "قزوین" ,
"كرمان" : "کرمان" , 
"كردستان" : "کردستان",
"كرمانشاه" : "کرمانشاه" , 
"كهگيلويه وبويراحمد" : "کهگیلویه و بویراحمد" ,
"گيلان" : "گیلان" , 
"مرکزي" : "مرکزی",
"يزد" : "یزد"
}
)
```

```python
province_list = [
    'کل کشور', 'آذربایجان شرقی', 'آذربایجان غربی', 'اردبیل', 'اصفهان',
    'البرز', 'ایلام', 'بوشهر', 'تهران', 'چهارمحال و بختیاری',
    'خراسان جنوبی', 'خراسان رضوی', 'خراسان شمالی', 'خوزستان', 'زنجان',
    'سمنان', 'سیستان و بلوچستان', 'فارس', 'قزوین', 'قم', 'کردستان',
    'کرمان', 'کرمانشاه', 'کهگیلویه و بویراحمد', 'گلستان', 'گیلان',
    'لرستان', 'مازندران', 'مرکزی', 'هرمزگان', 'همدان', 'یزد'
]

province_to_code = {name: idx for idx, name in enumerate(province_list)}
Dataset['Province_Code'] = Dataset['Province'].map(province_to_code)
cols = list(Dataset.columns)
cols.remove('Province_Code')
cols.insert(1, 'Province_Code')
Dataset = Dataset[cols]
Dataset = Dataset.dropna()
```

```python
output_file_name = '10_WorkShop_Input_Energy_By_Province.xlsx'
Dataset.to_excel(f"{Setting['Output_Path']}/{output_file_name}", index=False)
```
