```python
import pandas as pd
import yaml

# ---------------------------------
# Display settings
# ---------------------------------
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:,.2f}'.format)

# ------------------------------
# Load settings and initialize
# ------------------------------
with open('../../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

# ------------------------------
# Load Temperature Dataset
# ------------------------------
file_name = 'General.xlsx'
file_path = f"{Setting['Output_Path_General']}/{file_name}"
sheet_name = 'Weather_Province_Year'
Temp = pd.read_excel(file_path, sheet_name = sheet_name)

annual_stats = (
    Temp
    .groupby('Year')
    .apply(lambda g: pd.Series({
        # Area-weighted average temperature
        'Temprature': (g['TempAvg'] * g['Area']).sum() / g['Area'].sum(),
        # Total rainfall (sum across provinces)
        'Rainfall': g['AnnualRain'].sum(),
        # Area-weighted average sunshine hours
        'SunHours': (g['SunHours'] * g['Area']).sum() / g['Area'].sum(),
    }))
    .reset_index()
)


# ------------------------------
# Export Weather by year 
# ------------------------------
output_file_name = 'General.xlsx'
sheet_name= 'Weather_Year'
path = f"{Setting['Output_Path_General']}/{output_file_name}"
with pd.ExcelWriter(path, engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    annual_stats.to_excel(writer,sheet_name=sheet_name ,index=False)
```
