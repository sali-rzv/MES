```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T10 = pd.read_excel(file_path, sheet_name='T10', skiprows=1)
        T10 = T10.rename(columns={
        'سال':'Year',
        'سطح کد فعالیت':'Industry_Category_Code',
        'کد فعالیت':'Industry_Code',
        ' فعالیت':'Industry_Name',
        'اول فروردین':'Inventory_Value_Begining.total',
        'Unnamed: 5':'Inventory_Value_Begining.produced',
        'Unnamed: 6':'Inventory_Value_Begining.semi.produced',
        'Unnamed: 7':'Inventory_Value_Begining.non.transformed',
        'Unnamed: 8':'Inventory_Value_Begining.raw.material',
        'Unnamed: 9':'Inventory_Value_Begining.miscellaneous',
        'پایان اسفند':'Inventory_Value_Ending.total',
        'Unnamed: 11':'Inventory_Value_Ending.produced',
        'Unnamed: 12':'Inventory_Value_Ending.semi.produced',
        'Unnamed: 13':'Inventory_Value_Ending.non.transformed',
        'Unnamed: 14':'Inventory_Value_Ending.raw.material',
        'Unnamed: 15':'Inventory_Value_Ending.miscellaneous'
        })
        T10 = T10[T10['Year'] != 0]
        Data.append(T10)


    elif year == 1401:
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T10 = pd.read_excel(file_path, sheet_name='T10', skiprows=2)
        T10 = T10.rename(columns={
        'سال':'Year',
        'سطح کد فعالیت':'Industry_Category_Code',
        'کد فعالیت':'Industry_Code',
        ' فعالیت':'Industry_Name',
        'اول فروردین':'Inventory_Value_Begining.total',
        'Unnamed: 5':'Inventory_Value_Begining.produced',
        'Unnamed: 6':'Inventory_Value_Begining.semi.produced',
        'Unnamed: 7':'Inventory_Value_Begining.non.transformed',
        'Unnamed: 8':'Inventory_Value_Begining.raw.material',
        'Unnamed: 9':'Inventory_Value_Begining.miscellaneous',
        'پایان اسفند':'Inventory_Value_Ending.total',
        'Unnamed: 11':'Inventory_Value_Ending.produced',
        'Unnamed: 12':'Inventory_Value_Ending.semi.produced',
        'Unnamed: 13':'Inventory_Value_Ending.non.transformed',
        'Unnamed: 14':'Inventory_Value_Ending.raw.material',
        'Unnamed: 15':'Inventory_Value_Ending.miscellaneous'
        })
        T10 = T10[T10['Year'] != 0]
        Data.append(T10)

Dataset = pd.concat(Data, ignore_index=True)
Dataset = Dataset.dropna(subset=['Year'])
Dataset = Dataset.dropna(subset=['Industry_Category_Code'])
```

```python
output_file_name = '17_WorkShop_Inventory_By_Activity.xlsx'
Dataset.to_excel(f"{Setting['Output_Path']}/{output_file_name}", index=False)
```
