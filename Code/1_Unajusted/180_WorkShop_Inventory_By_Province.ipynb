```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T20 = pd.read_excel(file_path, sheet_name='T20', skiprows=1)
        T20 = T20.rename(columns={
        'سال':'Year',
        'استان':'Province',
        'اول فروردین':'Inventory_Value_Begining.total',
        'Unnamed: 3':'Inventory_Value_Begining.produced',
        'Unnamed: 4':'Inventory_Value_Begining.semi.produced',
        'Unnamed: 5':'Inventory_Value_Begining.non.transformed',
        'Unnamed: 6':'Inventory_Value_Begining.raw.material',
        'Unnamed: 7':'Inventory_Value_Begining.miscellaneous',
        'پایان اسفند':'Inventory_Value_Ending.total',
        'Unnamed: 9':'Inventory_Value_Ending.produced',
        'Unnamed: 10':'Inventory_Value_Ending.semi.produced',
        'Unnamed: 11':'Inventory_Value_Ending.non.transformed',
        'Unnamed: 12':'Inventory_Value_Ending.raw.material',
        'Unnamed: 13':'Inventory_Value_Ending.miscellaneous'
        })
        T20 = T20[T20['Year'] != 0]
        Data.append(T20)


    elif year == 1401:
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T20 = pd.read_excel(file_path, sheet_name='T20', skiprows=2)
        T20 = T20.rename(columns={
        'سال':'Year',
        'استان':'Province',
        'اول فروردین':'Inventory_Value_Begining.total',
        'Unnamed: 3':'Inventory_Value_Begining.produced',
        'Unnamed: 4':'Inventory_Value_Begining.semi.produced',
        'Unnamed: 5':'Inventory_Value_Begining.non.transformed',
        'Unnamed: 6':'Inventory_Value_Begining.raw.material',
        'Unnamed: 7':'Inventory_Value_Begining.miscellaneous',
        'پایان اسفند':'Inventory_Value_Ending.total',
        'Unnamed: 9':'Inventory_Value_Ending.produced',
        'Unnamed: 10':'Inventory_Value_Ending.semi.produced',
        'Unnamed: 11':'Inventory_Value_Ending.non.transformed',
        'Unnamed: 12':'Inventory_Value_Ending.raw.material',
        'Unnamed: 13':'Inventory_Value_Ending.miscellaneous'
        })
        T20 = T20[T20['Year'] != 0]
        Data.append(T20)

Dataset = pd.concat(Data, ignore_index=True)
Dataset = Dataset.dropna(subset=['Year'])
Dataset = Dataset.dropna(subset=['Province'])
```

```python
Dataset["Province"] = Dataset["Province"].replace({
"چهارمحال وبختيارئ" : "چهارمحال و بختیاری",
"اردبيل" : "اردبیل",
"ايلام" : "ایلام",
"آذربايجان شرقي" : "آذربایجان شرقی",
"آذربايجان غربي" : "آذربایجان غربی",
"خراسان جنوبي" : "خراسان جنوبی",
"خراسان رضوئ" : "خراسان رضوی",
"خراسان شمالي" : "خراسان شمالی" ,
"سيستان وبلوچستان" : "سیستان و بلوچستان" , 
"قزوين" : "قزوین" ,
"كرمان" : "کرمان" , 
"كردستان" : "کردستان",
"كرمانشاه" : "کرمانشاه" , 
"كهگيلويه وبويراحمد" : "کهگیلویه و بویراحمد" ,
"گيلان" : "گیلان" , 
"مرکزي" : "مرکزی",
"يزد" : "یزد"
})
```

```python
province_list = [
    'کل کشور', 'آذربایجان شرقی', 'آذربایجان غربی', 'اردبیل', 'اصفهان',
    'البرز', 'ایلام', 'بوشهر', 'تهران', 'چهارمحال و بختیاری',
    'خراسان جنوبی', 'خراسان رضوی', 'خراسان شمالی', 'خوزستان', 'زنجان',
    'سمنان', 'سیستان و بلوچستان', 'فارس', 'قزوین', 'قم', 'کردستان',
    'کرمان', 'کرمانشاه', 'کهگیلویه و بویراحمد', 'گلستان', 'گیلان',
    'لرستان', 'مازندران', 'مرکزی', 'هرمزگان', 'همدان', 'یزد'
]

province_to_code = {name: idx for idx, name in enumerate(province_list)}
Dataset['Province_Code'] = Dataset['Province'].map(province_to_code)
cols = list(Dataset.columns)
cols.remove('Province_Code')
cols.insert(1, 'Province_Code')
Dataset = Dataset[cols]
Dataset = Dataset.dropna(subset=['Year'])
```

```python
output_file_name = '180_WorkShop_Inventory_By_Province.xlsx'
Dataset.to_excel(f"{Setting['Output_Path_Unajusted']}/{output_file_name}", index=False)
```
