```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1399 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T17 = pd.read_excel(file_path, sheet_name='T17', skiprows=1)
        T17 = T17.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            ' جمع': 'Input_Services_Not_Industrial_Value',
            'حق العمل فروش      ' : 'Sale.Commission',
            '   اجاره ساختمان      ' : 'Rent.Building',
            '   اجاره ماشین آلات ' : 'Rent.Machine' ,
            '  ارتباطات و مخابرات  ' : 'Communication' , 
            'حمل و نقل' : 'Transportaion' ,
            ' حق بیمه پرداختی    ' : 'Insurance' ,
            '  خدمات حسابرسی و حقوقی     ' : 'Audit_Legal' ,
            'خدمات آموزشی' : 'Education' ,
            ' کارمزدهای پرداختی به بانک‌ها' : 'Bank_Fees',
            '  هزینه تحقیقات و آزمایشگاه' : 'Research',
            'هزینه تبلیغات، آگهی،نمایشگاه و مطبوعات' : 'Advertisement',
            ' حق ماموریت شاغلان  ' :'Mission',
            ' سایر' : 'ETC'
        })
        T17 = T17[T17['Year'] != 0]
        Data.append(T17)


    elif year == 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T17 = pd.read_excel(file_path, sheet_name='T17', skiprows=1)
        T17 = T17.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            ' جمع': 'Input_Services_Not_Industrial_Value',
            'حق‌العمل فروش      ' : 'Sale.Commission',
            '   اجاره ساختمان      ' : 'Rent.Building',
            '   اجاره ماشین‌آلات ' : 'Rent.Machine' ,
            '  ارتباطات و مخابرات  ' : 'Communication' , 
            'حمل‌ونقل' : 'Transportaion' ,
            ' حق بیمه پرداختی    ' : 'Insurance' ,
            '  خدمات حسابرسی و حقوقی     ' : 'Audit_Legal' ,
            'خدمات آموزشی' : 'Education' ,
            ' کارمزدهای پرداختی به بانک‌ها' : 'Bank_Fees',
            '  هزینه تحقیقات و آزمایشگاه' : 'Research',
            'هزینه تبلیغات، آگهی،نمایشگاه و مطبوعات' : 'Advertisement',
            ' حق مأموریت شاغلان  ' :'Mission',
            ' سایر' : 'ETC'
        })
        T17 = T17[T17['Year'] != 0]
        Data.append(T17)
    
    elif year == 1401 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T17 = pd.read_excel(file_path, sheet_name='T17', skiprows=2)
        T17 = T17.rename(columns={
            'سال':'Year',
            'استان' : 'Province' ,
            ' جمع': 'Input_Services_Not_Industrial_Value',
            'حق‌العمل فروش      ' : 'Sale.Commission',
            '   اجاره ساختمان      ' : 'Rent.Building',
            '   اجاره ماشین‌آلات ' : 'Rent.Machine' ,
            '  ارتباطات و مخابرات  ' : 'Communication' , 
            'حمل‌ونقل' : 'Transportaion' ,
            ' حق بیمه پرداختی    ' : 'Insurance' ,
            '  خدمات حسابرسی و حقوقی     ' : 'Audit_Legal' ,
            'خدمات آموزشی' : 'Education' ,
            ' کارمزدهای پرداختی به بانک‌ها' : 'Bank_Fees',
            '  هزینه تحقیقات و آزمایشگاه' : 'Research',
            'هزینه تبلیغات، آگهی،نمایشگاه و مطبوعات' : 'Advertisement',
            ' حق مأموریت شاغلان  ' :'Mission',
            ' سایر' : 'ETC'
        })
        T17 = T17[T17['Year'] != 0]
        Data.append(T17) 
    
Dataset = pd.concat(Data, ignore_index=True)
Dataset = Dataset.dropna(subset=['Year'])
```

```python
Dataset["Province"] = Dataset["Province"].replace({
"چهارمحال وبختيارئ" : "چهارمحال و بختیاری",
"اردبيل" : "اردبیل",
"ايلام" : "ایلام",
"آذربايجان شرقي" : "آذربایجان شرقی",
"آذربايجان غربي" : "آذربایجان غربی",
"خراسان جنوبي" : "خراسان جنوبی",
"خراسان رضوئ" : "خراسان رضوی",
"خراسان شمالي" : "خراسان شمالی" ,
"سيستان وبلوچستان" : "سیستان و بلوچستان" , 
"قزوين" : "قزوین" ,
"كرمان" : "کرمان" , 
"كردستان" : "کردستان",
"كرمانشاه" : "کرمانشاه" , 
"كهگيلويه وبويراحمد" : "کهگیلویه و بویراحمد" ,
"گيلان" : "گیلان" , 
"مرکزي" : "مرکزی",
"يزد" : "یزد"
}
)
```

```python
province_list = [
    'کل کشور', 'آذربایجان شرقی', 'آذربایجان غربی', 'اردبیل', 'اصفهان',
    'البرز', 'ایلام', 'بوشهر', 'تهران', 'چهارمحال و بختیاری',
    'خراسان جنوبی', 'خراسان رضوی', 'خراسان شمالی', 'خوزستان', 'زنجان',
    'سمنان', 'سیستان و بلوچستان', 'فارس', 'قزوین', 'قم', 'کردستان',
    'کرمان', 'کرمانشاه', 'کهگیلویه و بویراحمد', 'گلستان', 'گیلان',
    'لرستان', 'مازندران', 'مرکزی', 'هرمزگان', 'همدان', 'یزد'
]

province_to_code = {name: idx for idx, name in enumerate(province_list)}
Dataset['Province_Code'] = Dataset['Province'].map(province_to_code)
cols = list(Dataset.columns)
cols.remove('Province_Code')
cols.insert(1, 'Province_Code')
Dataset = Dataset[cols]
Dataset = Dataset.dropna(subset=['Year'])
```

```python
output_file_name = '150_WorkShop_Input_Not_Industrial_Service_By_Province.xlsx'
Dataset.to_excel(f"{Setting['Output_Path_Unajusted']}/{output_file_name}", index=False)
```
