```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1):
    if year <= 1399 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T15 = pd.read_excel(file_path, sheet_name='T15', skiprows=1)
        T15 = T15.rename(columns={
            'سال':'Year',
            'استان' : 'Province',
            'جمع' : 'Output_Value',
            'محصولات تولید شده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'ارزش غذای طبخ شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته شده' : 'Output_Value_Electricity',
            'آب فروخته شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
            'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
            'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به فروش رسیده اند' : 'Output_Value_speculate'
        })

        T15 = T15[(T15['Year'] != 0)]
        Data.append(T15)

    elif year == 1400 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T15 = pd.read_excel(file_path, sheet_name='T15', skiprows=1)
        T15 = T15.rename(columns={
            'سال':'Year',
            'استان' : 'Province',
            'جمع' : 'Output_Value',
            'محصولات تولیدشده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'ارزش غذای طبخ‌شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته‌شده' : 'Output_Value_Electricity',
            'آب فروخته‌شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
            'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
             'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به‌فروش‌رسیده‌اند' : 'Output_Value_speculate'
        })

        T15 = T15[(T15['Year'] != 0) & (T15['Year'].notna())]
        Data.append(T15)    
    
    
    elif year == 1401 :
        file_name = f"Y{year}_10AndMore.xlsx"
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T15 = pd.read_excel(file_path, sheet_name='T15', skiprows=2)
        T15 = T15.rename(columns={
            'سال':'Year',
            'استان' : 'Province',
            'جمع' : 'Output_Value',
            'محصولات تولیدشده' : 'Produced_Value' ,
            'ضایعات قابل فروش' : 'Wasted_Sold_Value',
            'ارزش غذای طبخ‌شده توسط کارگاه' : 'Output_Value_Baked_Food',
            'برق فروخته‌شده' : 'Output_Value_Electricity',
            'آب فروخته‌شده' : 'Output_Value_Water' ,
            'تغییرات موجودی انبار کالاهای در جریان ساخت' : 'Inventory_Change',
            'ساخت یا ایجاد و تعمیر اساسی اموال سرمایه ای توسط کارگاه' : 'Output_Value_Maintenance',
            'دریافتی بابت خدمات صنعتی ' : 'Output_Value_Receivable_Service',
            'تفاوت ارزش خرید و فروش کالاهایی که بدون تغییر شکل به‌فروش‌رسیده‌اند' : 'Output_Value_speculate'
        })

        T15 = T15[(T15['Year'] != 0)  & (T15['Year'].notna())]
        Data.append(T15)
    

Dataset = pd.concat(Data, ignore_index=True)
Dataset = Dataset.iloc[:,0:12]
```

```python
Dataset["Province"] = Dataset["Province"].replace({
"چهارمحال وبختيارئ" : "چهارمحال و بختیاری",
"اردبيل" : "اردبیل",
"ايلام" : "ایلام",
"آذربايجان شرقي" : "آذربایجان شرقی",
"آذربايجان غربي" : "آذربایجان غربی",
"خراسان جنوبي" : "خراسان جنوبی",
"خراسان رضوئ" : "خراسان رضوی",
"خراسان شمالي" : "خراسان شمالی" ,
"سيستان وبلوچستان" : "سیستان و بلوچستان" , 
"قزوين" : "قزوین" ,
"كرمان" : "کرمان" , 
"كردستان" : "کردستان",
"كرمانشاه" : "کرمانشاه" , 
"كهگيلويه وبويراحمد" : "کهگیلویه و بویراحمد" ,
"گيلان" : "گیلان" , 
"مرکزي" : "مرکزی",
"يزد" : "یزد"
}
)
```

```python
province_list = [
    'کل کشور', 'آذربایجان شرقی', 'آذربایجان غربی', 'اردبیل', 'اصفهان',
    'البرز', 'ایلام', 'بوشهر', 'تهران', 'چهارمحال و بختیاری',
    'خراسان جنوبی', 'خراسان رضوی', 'خراسان شمالی', 'خوزستان', 'زنجان',
    'سمنان', 'سیستان و بلوچستان', 'فارس', 'قزوین', 'قم', 'کردستان',
    'کرمان', 'کرمانشاه', 'کهگیلویه و بویراحمد', 'گلستان', 'گیلان',
    'لرستان', 'مازندران', 'مرکزی', 'هرمزگان', 'همدان', 'یزد'
]

province_to_code = {name: idx for idx, name in enumerate(province_list)}
Dataset['Province_Code'] = Dataset['Province'].map(province_to_code)
cols = list(Dataset.columns)
cols.remove('Province_Code')
cols.insert(1, 'Province_Code')
Dataset = Dataset[cols]

```

```python
output_file_name = '130_WorkShop_Output_By_Province.xlsx'
Dataset.to_excel(f"{Setting['Output_Path_Unajusted']}/{output_file_name}", index=False)
```
