```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1 ):
    if year <= 1399:
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T14 = pd.read_excel(file_path, sheet_name='T14', skiprows=1)
        T14 = T14.rename(columns={
            'سال':'Year',
            'استان':'Province',
            'جمع':'Input_Value',
            'مواد خام و اولیه، لوازم بسته بندی، ابزار و وسایل کار کم دوام' : 'Input_Value_Without_Energy',
            'Unnamed: 4' : 'Input_Value_without_Energy.Raw.Material',
            'Unnamed: 5' : 'Input_Value_without_Energy.Packing',
            'Unnamed: 6' : 'Input_Value_without_Energy.Low.Durable',
            'مواد مصرفی غذای طبخ شده توسط کارگاه' : 'Input_Value_Baked_Food',
            'لوازم مصرفی برای تعمیر اساسی اموال سرمایه ای توسط کارگاه' : 'Input_Value_Maintenance',
            'سوخت مصرف شده' : 'Input_Value_Fuel',
            'برق خریداری شده' : 'Input_Value_Electricity',
            'آب خریداری شده' : 'Input_Value_Water',
            'پرداختی بابت خدمات صنعتی ' : 'Input_Value_Payable_Service'
        })
        T14 = T14.iloc[2:]
        T14 = T14[T14['Year'] != 0] 
        Data.append(T14)
    
    
    elif  year  == 1400 :
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T14 = pd.read_excel(file_path, sheet_name='T14', skiprows=1)
        T14 = T14.rename(columns={
            'سال':'Year',
            'استان':'Province',
            'جمع':'Input_Value',
            'مواد خام و اولیه، لوازم بسته‌بندی، ابزار و وسایل کار کم‌دوام' : 'Input_Value_Without_Energy',
            'Unnamed: 4' : 'Input_Value_without_Energy.Raw.Material',
            'Unnamed: 5' : 'Input_Value_without_Energy.Packing',
            'Unnamed: 6' : 'Input_Value_without_Energy.Low.Durable',
            'مواد مصرفی غذایی طبخ‌شده توسط کارگاه' : 'Input_Value_Baked_Food',
            'لوازم مصرفی برای تعمیر اساسی اموال سرمایه‌ای توسط کارگاه' : 'Input_Value_Maintenance',
            'سوخت مصرف‌شده' : 'Input_Value_Fuel',
            'برق خریداری‌شده' : 'Input_Value_Electricity',
            'آب خریداری‌شده' : 'Input_Value_Water',
            'پرداختی بابت خدمات صنعتی ' : 'Input_Value_Payable_Service'
        })
        T14 = T14.iloc[2:]
        T14 = T14[T14['Year'] != 0] 
        Data.append(T14)
    
    elif  year ==1401 :
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f"{Setting['Raw_Path']}/{file_name}"
        T14 = pd.read_excel(file_path, sheet_name='T14', skiprows=2)
        T14 = T14.rename(columns={
            'سال':'Year',
            'استان':'Province',
            'جمع':'Input_Value',
            'مواد خام و اولیه، لوازم بسته‌بندی، ابزار و وسایل کار کم‌دوام' : 'Input_Value_Without_Energy',
            'Unnamed: 4' : 'Input_Value_without_Energy.Raw.Material',
            'Unnamed: 5' : 'Input_Value_without_Energy.Packing',
            'Unnamed: 6' : 'Input_Value_without_Energy.Low.Durable',
            'مواد مصرفی غذای طبخ‌شده توسط کارگاه' : 'Input_Value_Baked_Food',
            'لوازم مصرفی برای تعمیر اساسی اموال سرمایه‌ای توسط کارگاه' : 'Input_Value_Maintenance',
            'سوخت مصرف‌شده' : 'Input_Value_Fuel',
            'برق خریداری‌شده' : 'Input_Value_Electricity',
            'آب خریداری‌شده' : 'Input_Value_Water',
            'پرداختی بابت خدمات صنعتی ' : 'Input_Value_Payable_Service'
        })
        T14 = T14.iloc[2:]
        T14 = T14[T14['Year'] != 0] 
        Data.append(T14)
        
Dataset = pd.concat(Data, ignore_index=True)
```
