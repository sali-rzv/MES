```python
import pandas as pd
import yaml

pd.set_option('display.max_columns', None)

with open('../Settings.yaml', 'r') as file:
    Setting = yaml.safe_load(file)

Data = []

for year in range(Setting['Start_Year'], Setting['End_Year'] + 1 ):
    if year <= 1399 :
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f'{Setting['Raw_Path']}/{file_name}'
        T12 = pd.read_excel(file_path, sheet_name='T12', skiprows=1)
        T12 = T12.rename(columns={
                    'سال':'Year',
                    'استان':'Province',
                    'کل شاغلان': 'Employee.Total',
                    ' شاغلان تولیدی' : 'Manufacture_Employee.Total',
                    'Unnamed: 4':'Manufacture_Employee.Unskilled.worker',
                    'Unnamed: 5':'Manufacture_Employee.Skilled.worker',
                    'Unnamed: 6':'Manufacture_Employee.Technician',
                    'Unnamed: 7':'Manufacture_Employee.Engineer',
                    ' شاغلان غیرتولیدی':'Non_Manufacture_Employee.Total'
        })
        T12 = T12.iloc[1:]
        T12 = T12[T12['Year'] != 0] 
        Data.append(T12)

    elif year == 1400 :
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f'{Setting['Raw_Path']}/{file_name}'
        T12 = pd.read_excel(file_path, sheet_name='T12', skiprows=1)
        T12 = T12.rename(columns={
                    'سال':'Year',
                    'استان':'Province',
                    'کل شاغلان': 'Employee.Total',
                    ' شاغلان تولیدی' : 'Manufacture_Employee.Total',
                    'Unnamed: 4':'Manufacture_Employee.Unskilled.worker',
                    'Unnamed: 5':'Manufacture_Employee.Skilled.worker',
                    'Unnamed: 6':'Manufacture_Employee.Technician',
                    'Unnamed: 7':'Manufacture_Employee.Engineer',
                    ' شاغلان غیر تولیدی':'Non_Manufacture_Employee.Total'
        })
        T12 = T12.iloc[1:]
        T12 = T12[T12['Year'] != 0] 
        Data.append(T12)
    
    elif year == 1401 :
        file_name = f'Y{year}_10AndMore.xlsx'
        file_path = f'{Setting['Raw_Path']}/{file_name}'
        T12 = pd.read_excel(file_path, sheet_name='T12', skiprows=2)
        T12 = T12.rename(columns={
                    'سال':'Year',
                    'استان':'Province',
                    'کل شاغلان': 'Employee.Total',
                    ' شاغلان تولیدی' : 'Manufacture_Employee.Total',
                    'Unnamed: 4':'Manufacture_Employee.Unskilled.worker',
                    'Unnamed: 5':'Manufacture_Employee.Skilled.worker',
                    'Unnamed: 6':'Manufacture_Employee.Technician',
                    'Unnamed: 7':'Manufacture_Employee.Engineer',
                    ' شاغلان غیر تولیدی':'Non_Manufacture_Employee.Total'
        })
        T12 = T12.iloc[1:]
        T12 = T12[T12['Year'] != 0] 
        Data.append(T12)



Dataset = pd.concat(Data, ignore_index=True)  
Dataset.dropna(subset=['Year'], inplace=True)
```

```python
Dataset["Province"] = Dataset["Province"].replace({
"چهارمحال وبختيارئ" : "چهارمحال و بختیاری",
"اردبيل" : "اردبیل",
"ايلام" : "ایلام",
"آذربايجان شرقي" : "آذربایجان شرقی",
"آذربايجان غربي" : "آذربایجان غربی",
"خراسان جنوبي" : "خراسان جنوبی",
"خراسان رضوئ" : "خراسان رضوی",
"خراسان شمالي" : "خراسان شمالی" ,
"سيستان وبلوچستان" : "سیستان و بلوچستان" , 
"قزوين" : "قزوین" ,
"كرمان" : "کرمان" , 
"كردستان" : "کردستان",
"كرمانشاه" : "کرمانشاه" , 
"كهگيلويه وبويراحمد" : "کهگیلویه و بویراحمد" ,
"گيلان" : "گیلان" , 
"مرکزي" : "مرکزی",
"يزد" : "یزد"
}
)
```

```python
province_list = [
    'کل کشور', 'آذربایجان شرقی', 'آذربایجان غربی', 'اردبیل', 'اصفهان',
    'البرز', 'ایلام', 'بوشهر', 'تهران', 'چهارمحال و بختیاری',
    'خراسان جنوبی', 'خراسان رضوی', 'خراسان شمالی', 'خوزستان', 'زنجان',
    'سمنان', 'سیستان و بلوچستان', 'فارس', 'قزوین', 'قم', 'کردستان',
    'کرمان', 'کرمانشاه', 'کهگیلویه و بویراحمد', 'گلستان', 'گیلان',
    'لرستان', 'مازندران', 'مرکزی', 'هرمزگان', 'همدان', 'یزد'
]
province_to_code = {name: idx for idx, name in enumerate(province_list)}
Dataset['Province_Code'] = Dataset['Province'].map(province_to_code)
cols = list(Dataset.columns)
cols.remove('Province_Code')
cols.insert(1, 'Province_Code')
Dataset = Dataset[cols]

```

```python
output_file_name = '12_WorkShop_Employee_By_Province.xlsx'
Dataset.to_excel(f"{Setting['Output_Path']}/{output_file_name}", index=False)
```
